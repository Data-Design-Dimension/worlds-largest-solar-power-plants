# Defining 'Big Power' in the 20 Largest Solar Power Plants

Analysis and data visualization by Kathryn Hurchla for Data Design Dimension

```python
from matplotlib import cbook
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from mpl_toolkits.axes_grid1 import ImageGrid
import pandas as pd
import numpy as np
```

```python
# load the entire Excel data file into a pandas dataframe, which contains only one sheet and just a few columns
df = pd.read_excel("./data/20 Largest Solar Power Plants.xlsx")
# display the dataframe
df.head()
```

```python
# calculate MWp/Acre and MWp/SqMi and MWp/SqKm and append each as columns to the dataframe
df["MWp/Acre"] = df["MWp (Megawatts Peak Capacity)"] / df["Acres"]
df["MWp/Sq Mi"] = df["MWp (Megawatts Peak Capacity)"] / df["Sq Mi"]
df["MWp/Sq Km"] = df["MWp (Megawatts Peak Capacity)"] / df["Sq Km"]
df
```

## Among the Top 20 Solar Plants

As we look at the top 20 plants as measured by peak capacity each generates, in megawatts, in the scatterplot graph below, a trend appears to show that a larger plant, in acreage, produces a higher peak capacity of solar energy.

```python
# plot relationship between peak capacity and acreage

x = df["MWp (Megawatts Peak Capacity)"]
y = df["Acres"]
# use a Solarize Light stylesheet from matplotlib
with plt.style.context('Solarize_Light2'):
    # use a scatter plot
    plt.scatter(x, y)
    plt.title('Does more acreage mean a higher peak capacity?\nTop 20 Solar Plants shown 2016-2020', pad=20)
    plt.xlabel('Megawatts Peak Capacity (MWp)', fontsize=14)
    plt.ylabel('Acres', fontsize=14)
    # add a trend line using numpy
    z = np.polyfit(x, y, 1)
    p = np.poly1d(z)
    plt.plot(x,p(x), color="gold") # color the trend line gold #"y--") # color the trend line yellow
    # add a source footer
    plt.annotate("Source Article: SolarPower.guide | Data Source: Wikipedia\nDesign: Data Design Dimension (c) 2021", (0,0), (0, -50), xycoords='axes fraction', textcoords='offset points', va='top', color="gray", size=11.5)
        
plt.show()
```

## Among the Top Plants Limited to 2020

The top plants have been measured by peak wattage ranging in years from 2016 to 2020, as we have seen in our data table above. Graphing only the most recent year available, 2020, only four plants among the same top 20 were appear as recorded with these highs in 2020. 

Although the trend appears to be reversed towards less acreage and lower wattage, the number of plants are not sufficient to make any conclusions from within this single year. If a full dataset of 2020 recordings was available, more could be seen about how acreage and peak capacity related to one another in that year.

```python
# limit plot to only 2020 year plant records, and plot relationship between peak capacity and acreage

df_2020 = df.loc[(df.Year == 2020)]

x = df_2020["MWp (Megawatts Peak Capacity)"]
y = df_2020["Acres"]
# use a Solarize Light stylesheet from matplotlib
with plt.style.context('Solarize_Light2'):
    # use a scatter plot
    plt.scatter(x, y)
    plt.title('Does more acreage mean a higher peak capacity?\nTop Solar Plants in 2020 shown', pad=20)
    plt.xlabel('Megawatts Peak Capacity (MWp)', fontsize=14)
    plt.ylabel('Acres of Solar Power Plant', fontsize=14)
    # add a trend line using numpy
    z = np.polyfit(x, y, 1)
    p = np.poly1d(z)
    plt.plot(x,p(x), color="gold") # color the trend line gold #"y--") # color the trend line yellow    
    # add a source footer
    plt.annotate("Source Article: SolarPower.guide | Data Source: Wikipedia\nDesign: Data Design Dimension (c) 2021", (0,0), (0, -50), xycoords='axes fraction', textcoords='offset points', va='top', color="gray", size=11.5)
     
plt.show()
```

## Did we make more power per acre in 2020 than in 2016?

The real question becomes, are the top power plants become more efficient, or are just bigger?

To see if a time based trend has developed over the years these peak capacities were recorded, the Megawatts Peak Capacity of each power plant was divided by the plant's number of Acres, for a MWp by Acre measurement.

A more modest, but more meaningful, trend shows that the top 20 power plants were indeed making an increasingly higher peak capacity of solar energy per acre over this six year timespan. 

```python
# plot over time the relationship between peak capacity and acreage (MWp/Acre by Year)

x = df["Year"]
y = df["MWp/Acre"]
# use a Solarize Light stylesheet from matplotlib
with plt.style.context('Solarize_Light2'):
    # use a scatter plot
    plt.scatter(x, y)
    plt.title('Are We Making More Peak Capacity Per Acre?\nTop 20 Solar Plants shown 2016-2020', pad=20)
    plt.xlabel('Megawatts Peak Capacity by Acre Over Time', fontsize=14)
    plt.ylabel('MWp by Acre', fontsize=14)
    # add a trend line using numpy
    z = np.polyfit(x, y, 1)
    p = np.poly1d(z)
    plt.plot(x,p(x), color="gold") # color the trend line gold #"y--") # color the trend line yellow
    # add a source footer
    plt.annotate("Source Article: SolarPower.guide | Data Source: Wikipedia\nDesign: Data Design Dimension (c) 2021", (0,0), (0, -50), xycoords='axes fraction', textcoords='offset points', va='top', color="gray", size=11.5)
    
plt.show()
```
